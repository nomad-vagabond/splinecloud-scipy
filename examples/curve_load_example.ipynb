```python
from splinecloud_scipy import load_spline
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
curve_id = 'spl_K5t56P5bormJ'
spline = load_spline(curve_id)
```

```python
X = np.linspace(0, 20, 100)
Y = [spline.eval(x) for x in X]
```

```python
plt.plot(X,Y)
plt.grid()
plt.show()
```

```python
Y = [spline.eval(x, extrapolate=True) for x in X]
```

```python
plt.plot(X,Y)
plt.grid()
plt.show()
```

```python
columns, table = spline.load_data()
x_data, y_data = table.T
```

```python
plt.plot(X,Y)
plt.plot(x_data, y_data, 'o')
plt.grid()
plt.xlabel(columns[0])
plt.ylabel(columns[1])
plt.show()
```
