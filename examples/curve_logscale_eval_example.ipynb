```python
import requests, json
import numpy as np
import splinecloud_scipy as scsp
import matplotlib.pyplot as plt
%matplotlib inline
```

## Evaluate curve with log Y scale

```python
spline = scsp.load_spline('spl_Z5PEJhGIKziu')
```

```python
X = np.linspace(1900, 2020, 50)
Y = spline.eval(X, extrapolate=True)
Y
```

```python
spline.eval(1995)
```

```python
C, D = spline.load_data()
```

```python
plt.figure()
plt.plot(X,Y)
plt.scatter(D[:, 0], D[:, 1])
plt.title("Cl vs AoA (degres)")
plt.xlabel("AoA")
plt.grid()
plt.yscale("log")
plt.show()
```
