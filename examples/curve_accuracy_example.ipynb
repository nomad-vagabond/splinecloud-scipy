```python
from splinecloud_scipy import load_spline, load_subset
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
spline = load_spline('spl_EJ7JmWRj3oCu')
columns, table  = load_subset("sbt_CGysxfMtdbTL")
```

```python
X = np.linspace(0, 16, 100)
Y = [spline.eval(x) for x in X]
x_data, y_data = table.T
```

```python
plt.plot(X,Y)
plt.plot(x_data, y_data, "o")
plt.grid()
plt.show()
```
